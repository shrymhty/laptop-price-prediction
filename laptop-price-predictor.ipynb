```python
def convert_tb_to_gb(value):
    if 'TB' in value:
        tb_value = int(value.replace('TB', ''))
        gb_value = tb_value * 1024
        return f'{gb_value}GB'
    else:
        return value
```

```python
df['ROM'] = df['ROM'].apply(convert_tb_to_gb)
```

```python
df['ROM'] = df['ROM'].str.replace('GB','')
```
